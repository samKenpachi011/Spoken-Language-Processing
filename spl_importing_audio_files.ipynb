## -WE'll start by importing one audio file

 **Seeing what it looks line in byte form using the build-in readframes() method**
 
 **Change from bytes to integers**
 
 **Then we'll find the timesstamps for each integer**
 
 **We'll add another audio and compare both audios**
 
 **Finally we'll use matplotlib to visualize the timestamps**


```python
#Import Wave
import wave
# Create audio file wave object
good_morning = wave.open('sounds/r3_goodMorning.wav', 'r')
```

```python
# Read all frames from wave object 
signal_gm = good_morning.readframes(-1)
# View first 10
print(signal_gm[:10])
```

```python
#Change the signal_gm to the right data type for visualizing 
#dtype defaults to float in np.frombuffer(), use int16
# Convert good morning audio bytes to integers
import numpy as np
soundwave_gm = np.frombuffer(signal_gm, dtype='int16')
# View the first 10 sound wave values
print(soundwave_gm[:10])
```

```python
# Get the sound wave frame rate
framerate_gm = good_morning.getframerate()
# Find the sound wave time
stampstime_gm = np.linspace(
	start=0,
    stop=len(soundwave_gm)/framerate_gm,
	num=len(soundwave_gm)
)
# Print the first 10 timestamps
print(stampstime_gm[:10])
```

```python
#We'll use MatPlotLib to vizualise the timestamps
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(stampstime_gm,soundwave_gm,alpha=2)
plt.ylabel('Amplitude')
plt.xlabel('Time (seconds)')
plt.show()
```

## Now lets add another audio file and plot them to see how they compare

```python
#now lets add a good afternoon wav and compare both
# Create audio file wave object
good_afternoon = wave.open('sounds/r1_goodAfternoon.wav', 'r')
# Read all frames from wave object 
signal_ga = good_afternoon.readframes(-1)
# View first 10
print(signal_ga[:10])
#convert to the right datatype
soundwave_ga = np.frombuffer(signal_ga, dtype='int16')
# View the first 10 sound wave values
print(soundwave_ga[:10])
# Get the sound wave frame rate
framerate_ga = good_afternoon.getframerate()
# Find the sound wave time
stampstime_ga = np.linspace(
	start=0,
  stop=len(soundwave_ga)/framerate_ga,
	num=len(soundwave_ga)
)
# Print the first 10 timestamps
print(stampstime_ga[:10])
```

```python
plt.title('Good Afternoon vs. Good Morning')
plt.ylabel('Amplitude')
plt.xlabel('Time (seconds)')
# Add the Good Afternoon data to the plot
plt.plot(soundwave_ga, label='Good Afternoon')
# Add the Good Morning data to the plot
plt.plot(soundwave_gm, label='Good Morning',
   # Set the alpha variable to 0.5
alpha=0.5)
plt.legend()
plt.show()
```
